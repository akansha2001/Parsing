Defining Nodes

```python
# define node for Tree
class NodeT:
    def __init__(self,val):
        self.val = val
        self.c0 = None
        self.c1 = None
        self.c2 = None
        self.c3 = None
        self.c4 = None
# define node for Linked List
class NodeLL:
    def __init__(self,val):
        self.val = val
        self.next = None
```

Class for Parsing

```python
class Parsing:
    def __init__(self):
        pass
    def grammar(self,add,rw = '=',seq = ',',orop = '|'):
        '''
        add: address for the grammar file
        
        Split parameters:
        rw: replace with operator (default =)
        seq: operator to denote sequence (default ,)
        orop: operator to denote OR (default |)
        '''
        f = open(add,'r')
        lines = f.readlines()

        # dictionary for grammar and rules
        gram = {}
        rules = {}
        c = 1

        for line in lines:

            line = line.strip()
            rule = line.split(rw)
            x = rule[0]
            y = rule[1]
            gram[x] = y.split(orop)
            rules[x] = c
            c += 1


            for i in range(len(gram[x])):
                gram[x][i] = gram[x][i].split(seq)
                
        self.lines = lines                
        self.gram = gram
        self.rules = rules

    def readSentence(self):
        # read sentence 

        s = input("Enter a sentence:\n").split()

        start = NodeLL('Nada') # dummy word 
        n = start
        
        for w in s:
            n.next = NodeLL(w)
            n = n.next

        # drop the dummy word
        start = start.next
        
        # initialize marker
        self.s = start
        
    def parse(self,parent,marker):
        '''
        parent: current parent node of parse tree
        marker: denotes marked word in linked list
        '''
        
        if marker == None: # end of sentence
            if self.gram.get(parent.val)!=None: # expandable: sentence is incomplete according to grammar
                print("The sentence is incomplete")
                return False,marker
            else: # the sentence is complete
                return True,marker  

        elif parent.val == marker.val: # match
            print("Comparing marked word: ",marker.val," to node ",parent.val," ...match")
            return True,marker.next

        else: # no match   

            if self.gram.get(parent.val)!=None: # expandable
                print("Rule no. ",self.rules[parent.val]," ",self.lines[self.rules[parent.val]-1].strip()," comparing marked word: ",marker.val," to node ",parent.val," ...no match")

                for lis1 in self.gram[parent.val]:

                    # expand
                    if len(lis1)>=1:
                        parent.c0 = NodeT(lis1[0])
                    else:
                        parent.c0 = None

                    if len(lis1)>=2:
                        parent.c1 = NodeT(lis1[1])
                    else:
                        parent.c1 = None

                    if len(lis1)>=3:
                        parent.c2 = NodeT(lis1[2])
                    else:
                        parent.c2 = None

                    if len(lis1)>=4:
                        parent.c3 = NodeT(lis1[3])
                    else:
                        parent.c3 = None

                    if len(lis1)>=5:
                        parent.c4 = NodeT(lis1[4])
                    else:
                        parent.c4 = None



                    # node wise check
                    if parent.c0 != None:    
                        parsible0,pos0 = self.parse(parent.c0,marker)
                        if parsible0: 
                            if parent.c1 != None:
                                parsible1,pos1 = self.parse(parent.c1,pos0)
                                if parsible1:
                                    if parent.c2 != None:
                                        parsible2,pos2 = self.parse(parent.c2,pos1)
                                        if parsible2:
                                            if parent.c3 != None:
                                                parsible3,pos3 = self.parse(parent.c3,pos2)
                                                if parsible3:
                                                    if parent.c4 != None:
                                                        parsible4,pos4 = self.parse(parent.c4,pos3)
                                                        if parsible4:
                                                            return True,pos4
                                                        else:
                                                            continue
                                                    else:
                                                        return parsible3,pos3
                                                else:
                                                    continue
                                            else:
                                                return parsible2,pos2 
                                        else:       
                                            continue
                                    else:
                                        return parsible1,pos1
                                else: 
                                    continue
                            else:
                                return parsible0,pos0
                        else:
                            continue
                    else: 
                        return False,marker  

                return False,marker


            else: # no match and not expandable

                print("Comparing marked word: ",marker.val," to node ",parent.val," ... no match")
                return False,marker
```

Parsing with one set of grammar rules

```python
# object creation
obj = Parsing()
# initialize grammar 
obj.grammar('grammar.txt',rw = '=',seq = ',',orop = '|')
# read sentence
obj.readSentence()

print('\n')

# function call to initialize root node
S = NodeT('S')

par,pos = obj.parse(S,obj.s)

if par == True:
    print('\nThe sentence can been parsed successfully')
else:
    print('\nThe sentence can not be parsed')
```
